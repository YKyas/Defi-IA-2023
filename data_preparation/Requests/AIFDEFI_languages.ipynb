```python
# import pandas module 
import pandas as pd 
    
# making dataframe 
df = pd.read_csv("test_set.csv") 
   
# output the dataframe
print(df)
```

```python
d_language=df['language'].value_counts()
d_language = d_language.to_frame(name='language')
d_language['percent'] = (d_language['language']/
                    d_language['language'].sum())*100
d_language
```

```python
d_percentage_l=d_language['percent'].round()
d_percentage_l=d_percentage_l.to_frame(name='percentage')
d_percentage_l=d_percentage_l['percentage'].astype('int')
d_percentage_l= d_percentage_l.to_dict()
d_percentage_l
```

```python
c=102
list_requests=[]
for lang in d_percentage_l.keys():
    for i in range(d_percentage_l[lang]):
        name = 'yaa-'+str(c)+'-avatar'
        r = requests.post(path(f'avatars/{user_id}/{name}'))
        params = {
            "avatar_name": name,
            "language": lang,
            "city": random.choice(cities),
            "date": random.choice(dates),
            "mobile": random.choice(mobile)
            }
        list_requests.append(requests.get(path(f"pricing/{user_id}"), params=params))
        c+=1

```

```python
len(list_requests)
```

```python
pricing_requests = []

for r in list_requests:
    pricing_requests.append(
        pd.DataFrame(r.json()['prices']).assign(**r.json()['request'])
    )

pricing_requests = pd.concat(pricing_requests)
pricing_requests
```
