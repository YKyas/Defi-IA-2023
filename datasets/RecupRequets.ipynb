```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df_date1 = pd.read_csv(io.BytesIO(uploaded['pricing_requests_500randomreq_date1.csv']))
df_date22 = pd.read_csv(io.BytesIO(uploaded['455_pricing_requests_date22.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```python
df_date1 = df_date1.drop(df_date1.columns[[0]], axis=1)
```

```python
df_date22
```

```python
total_df= []
part1 = df_date22['request']
part2 = df_date22['prices']
for i in range(len(df_date22)):
  part1_dict = eval(df_date22['request'].iloc[i])
  part2_dict = eval(df_date22['prices'].iloc[i])
  
  #creating the request specifications dataframe
  for key in part1_dict.keys():
    part1_dict[key] = [part1_dict[key]]
  part1_df = pd.DataFrame.from_dict(part1_dict)
  
  #creating the results dataframe
  results = []
  for k in range(len(part2_dict)):
    temp_dict = part2_dict[k]
    for key in temp_dict.keys():
      temp_dict[key] = [temp_dict[key]]
    df_from_temp_dict = pd.DataFrame.from_dict(temp_dict)
    results.append(df_from_temp_dict)
  part2_df = pd.concat(results)
  total_df.append(part2_df.merge(part1_df, how='cross'))
```

```python
df_date22= pd.concat(total_df)
```

```python
df_date22
```

```python
#save the csv
pd.concat(total_df).to_csv('455_pricing_requests_date22_unzipped.csv')
```
